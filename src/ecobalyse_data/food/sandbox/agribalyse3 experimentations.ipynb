## Est ce qu'il y a des "consumption mix" sans "national average" en parents ?

```python
def parent_exchanges_with_kw(activity, keyword, check_parent = True):
    if check_parent:
        kw_exchanges = [exchange.input._data["name"] for exchange in activity.technosphere() if keyword in exchange.input._data["name"]]
           
    if not check_parent:  
        kw_exchanges = [exchange.input._data["name"] for exchange in activity.upstream() if keyword in exchange.input._data["name"]]
                
    return kw_exchanges
```

```python
search_results = list(agb.search("consumption mix",limit = 2000))   

cons_mix_avec_nat_average = []
kw = "national average"

data = []
for activity in search_results:        
    exch = parent_exchanges_with_kw(activity, kw)
    if len(exch) > 0:
        print(f"Activity : {activity}")
        print(f"   {kw} : {exch}")
        for ex in exch:
            data.append([activity["name"], ex])

    
df = pd.DataFrame(data, columns = ["activity","sub_activity" ])
df.to_excel("cons_mix_with_nat_avg.xlsx")
        
```
